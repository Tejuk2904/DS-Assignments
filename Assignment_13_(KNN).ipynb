```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.neighbors import  KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, KFold, cross_val_score

import warnings
warnings.filterwarnings('ignore')
```

## **KNN Model Building**

```python
x = glass.iloc[:, 0:9]
y = glass.iloc[:, -1]
```

```python
x
```

```python
y
```

```python
model = KNeighborsClassifier(n_neighbors = 12)
model.fit(x,y)
```

```python
kfold = KFold(n_splits = 8)
result = cross_val_score (model, x, y, cv = kfold)
result
```

```python
result.mean()
```

## Grid search for algorithm Tuning

```python
n_neighbors = list(range(1,33))
para_grid = dict(n_neighbors = n_neighbors)
```

```python
model = KNeighborsClassifier()
grid = GridSearchCV(estimator = model , param_grid = para_grid)
grid.fit(x,y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

## Visualizing the CV results

```python
k_range = range(1,57)
k_scores = []

for k in k_range:
  knn = KNeighborsClassifier(n_neighbors = k)
  scores = cross_val_score(knn, x, y, cv=4)
  k_scores.append(scores.mean())


plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross_validated_Accuracy')
plt.show()
```

## **KNN Model Building**

```python
X = zoo.iloc[:, 1:17]
Y = zoo.iloc[:, -1]
X
```

```python
Y
```

```python
model2 = KNeighborsClassifier(n_neighbors = 17)
model2.fit(X,Y)
```

```python
kfold2 = KFold(n_splits = 12)
result2 = cross_val_score(model2, X, Y, cv=kfold2)
result2
```

```python
result2.mean()
```

## Grid Search Algorithm Tuning

```python
n_neighbors2 = list(range(1,40))
para_grid2 = dict(n_neighbors = n_neighbors2)
```

```python
model2 = KNeighborsClassifier()
grid2 = GridSearchCV(estimator = model2, param_grid = para_grid2 )
grid2.fit(X, Y)
```

```python
print(grid2.best_score_)
print(grid2.best_params_)
```

```python
k_range = range(1,35)
k_scores = []

for k in k_range:
  knn = KNeighborsClassifier(n_neighbors = k)
  scores = cross_val_score(knn , X, Y, cv=4)
  k_scores.append(scores.mean())


plt.figure(figsize=(16,6))
plt.plot(k_range, k_scores)
plt.xlabel('Value of k for KNN')
plt.ylabel('Cross_validated Accuracy')
plt.show()
```
