## **Build a recommender system by using cosine simillarties score (book1.csv)**

```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book=pd.read_csv("/content/book(1).csv",encoding = 'ISO-8859-1')
book
```

```python
books = book.iloc[:, 1:]
books
```

```python
# sort by user IDs
books.sort_values(['User.ID'])
```

```python
# Renaming the column
books = books.rename({'User.ID':'userId','Book.Title':'Title','Book.Rating':'Rating'}, axis=1)
books
```

```python
books.shape
```

```python
books.nunique()
```

```python
books['Rating'].value_counts()
```

```python
# converting long data into wide data using pivot table
books_2 = books.pivot_table(index = 'userId',
                            columns = 'Title',
                            values = 'Rating').reset_index(drop=True)
books_2
```

```python
# Replacing the index values by unique UserId
books_2.index = books.userId.unique()
books_2
```

```python
# impute those NaNs with 0 values
books_2.fillna(0,inplace=True)
books_2
```

```python
# Calculating Cosine Similarity between Users on array data
user_sim = 1 - pairwise_distances(books_2.values, metric = 'cosine')
user_sim
```

```python
# Store results in dataframe
user_sim_2 = pd.DataFrame(user_sim)
user_sim_2
```

```python
# Set the index and columns names to UserId
user_sim_2.index = books['userId'].unique()
user_sim_2.columns = books['userId'].unique()
user_sim_2
```

```python
# null diagonal values
np.fill_diagonal(user_sim, 0)
user_sim_2
```

```python
# most similar users
user_sim_2.idxmax(axis=1)
```
